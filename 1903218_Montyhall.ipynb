```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sticks Function

```python
def Stick():
        correct = np.random.randint(1,6)
        guess = np.random.randint(1,6)
        
        if correct == guess:
            return 1
        else:
            return 0
```

## Switch function 

```python
def Switch():

    all_possible_options = [1,2,3,4,5]    
    correct = np.random.choice(all_possible_options)
    guess = np.random.choice(all_possible_options)

    revealing_options = [1,2,3,4,5]
       
    
    revealing_options.remove(correct)
    if(correct != guess):
        revealing_options.remove(guess)   
    
    
    revealed_door = np.random.choice(revealing_options)
    # revealing_options.remove(revealed_door)
    # revealed_door2 = np.random.choice(revealing_options)

    second_choice_possiblity = [1,2,3,4,5]

    second_choice_possiblity.remove(revealed_door)
    # second_choice_possiblity.remove(revealed_door2)
    second_choice_possiblity.remove(guess)

    New_Guess = np.random.choice(second_choice_possiblity)

    # print("  Correct: ",correct ,"  guess: ", guess ,"  revealing options: ",revealing_options,"   revealed door: ",revealed_door ," ", revealed_door2 ,"  New Guess: " , New_Guess)

    if (New_Guess == correct):
        return 1
    else:
        return 0 
    
    
    
```

```python
num_of_iterartion = 1000
Mean = []
attempt_counter = []


record_switch = []

for i in range(1,num_of_iterartion):
    for j in (1,i):
        record_switch.append(Switch())
            
    Mean.append(np.mean(record_switch))
    attempt_counter.append(i)

plt.plot(attempt_counter,Mean , 'r')
plt.show()

    
    
```
